```python
all_on = np.ones((1024, 1024), dtype = np.uint8)*255
all_off = np.zeros((1024, 1024), dtype = np.uint8)
```

```python
grid_stims = draw_all_grid_rects(n_cols=12, n_rows=9)
```

```python
util_stim = np.append(all_on, all_off, axis = 2)
util_stim.shape
```

```python
checker_board = np.zeros((1024, 1024), dtype = np.uint8)
for i in np.arange(0, 108, 2):
    checker_board += grid_stims[:,:,i]
plt.imshow(checker_board)
    
```

```python
util_stims = np.stack([all_on, all_off, checker_board], axis=-1)
util_stims.shape
```

```python
sequence_dict = {}
sequence_dict['on'] = np.array([0])
sequence_dict['off'] = np.array([1])
sequence_dict['checker_board'] = np.array([2])
```

```python
utility_ss = StimSequenceSet(util_stims, 'utility_stims', sequence_dict)
```

```python
import pickle
with open(f'utility_stims.pickle', 'wb') as file:
    pickle.dump(utility_ss, file) 
```

```python
def get_summed_image(image_sequence, index_list):
    summed_image = np.zeros(image_sequence.shape[:-1], dtype=np.uint8)
    for i in index_list:
        summed_image += image_sequence[:,:,i]
    plt.imshow(summed_image)
```
